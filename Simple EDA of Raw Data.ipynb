# Basic Data Visualization

This file have some basic visulaization of data.
To check the relation between given numeric terms.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
%matplotlib inline
```

```python
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
credits.columns = ['id','tittle','cast','crew']
credits.head(5)
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
```

```python
movies.head(5)
```

```python
movies = movies.merge(credits, on='id')
```

```python
movies.head(5)
```

```python
plt.figure(figsize=(12,8))
sns.distplot(movies['vote_average'],kde=False,bins=30, color='darkblue')
```

```python
#sns.swarmplot(x="original_language", y="vote_average", data=movies)
```

```python
sns.jointplot(y='vote_count',x='vote_average',data=movies,kind='scatter', height=10)
```

```python
p=sns.jointplot(x='popularity',y='vote_count',data=movies,kind='reg',height=10, color='goldenrod')
p.ax_marg_y.set_ylim(0,5000)
p.ax_marg_x.set_xlim(10,100)
```
